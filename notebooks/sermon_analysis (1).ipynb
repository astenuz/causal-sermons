```python
from scipy.stats import ttest_ind

def hte_of_column(df, column_name, outcome):
  quantiles = df[column_name].quantile(q=0.5), df[column_name].quantile(q=0.5)
  r_1 = df[df[column_name] < quantiles[0]][outcome]
  r_2 = df[df[column_name] >= quantiles[1]][outcome]
  r_1_sd = round(r_1.std(), 3)
  r_1_mean = round(r_1.mean(), 3)
  r_2_sd = round(r_2.std(), 3)
  r_2_mean = round(r_2.mean(), 3)
  test_result = ttest_ind(r_1, r_2)

  print(f"Outcome: {outcome}")
  print(f"Column name: {column_name}")
  print(f'Lowest quartile standard deviation: {r_1_sd}')
  print(f'Lowest quartile mean: {r_1_mean}')
  print(f'Upper quartile standard deviation: {r_2_sd}')
  print(f'Upper quartile mean: {r_2_mean}')
  print(f'P-value? {test_result[1]}')
```

```python
## curious about HTE wrt lower income vs higher income and percent_adherents
#mean_income = sermons['mean_'].mean()
hte_of_column(sermon_filt, 'percent_adherents', 'trump_minus_clinton_tau')
hte_of_column(sermon_filt, 'percent_adherents', 'portion_voted_tau')
```
